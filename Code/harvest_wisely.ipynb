```python
import pandas as pd

df = pd.read_excel("../Data/RAMLDB v4.65 (assessment data only).xlsx", sheet_name="timeseries.1")
df = df.query("tsid == 'BdivBmsypref-dimensionless'")
df = df.query("~tsvalue.isna()")
df
```

```python
import matplotlib.pyplot as plt

filtered = df.query("assessid=='WGWIDE-HERRNORSS-1988-2020-ICESIMP2021-2'")

# Create the figure with specified DPI
plt.figure(figsize=(12, 8), dpi=300)

plt.plot(filtered["tsyear"], filtered["tsvalue"], linewidth=3)
plt.axhline(y=1, color='r', linestyle='-')

plt.title(f'MSY over time for {filtered.head(1)["stocklong"].values[0]}')
plt.xlabel('Year', fontsize=16)
plt.ylabel('MSY', fontsize=16)
plt.grid(True)

# Adjust layout to make sure everything fits well within the figure
plt.tight_layout(rect=[0, 0, 1, 0.85])

# Set the font to Avenir
plt.rcParams['font.family'] = 'Avenir'

# Set the font size for tick labels
plt.tick_params(axis='both', which='major', labelsize=14)

# Save the figure as an SVG file
plt.savefig('../Figures/MSY_example.svg', format='svg')

plt.show()
```
