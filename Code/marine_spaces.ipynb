```python
import geopandas as gpd
from shapely.ops import unary_union
from tqdm import tqdm

# Filter the DataFrame for non-Marine Protected Area
filtered_df = WDPA_df[~WDPA_df['DESIG_ENG'].isin(marine_designations)]

# Sort by STATUS_YR to ensure chronological order
filtered_df = filtered_df.sort_values(by='STATUS_YR')

# Initialize an empty GeoSeries to keep track of cumulative union of geometries
cumulative_union = gpd.GeoSeries()

# List to store the results
yearly_areas = []

# Process each year individually
for year in tqdm(filtered_df['STATUS_YR'].unique()):
    # Get the polygons for the current year
    yearly_polygons = filtered_df[filtered_df['STATUS_YR'] == year].geometry
    
    # Union the current year's polygons with the cumulative union
    cumulative_union = unary_union([*cumulative_union, *yearly_polygons])
    
    # Calculate the total area of the current cumulative union
    total_area = cumulative_union.area
    
    # Append the result to the list
    yearly_areas.append({'YEAR': year, 'TOTAL_AREA': total_area})

# Convert the results into a DataFrame
cumulative_sum_df = pd.DataFrame(yearly_areas)

# Display the resulting DataFrame
cumulative_sum_df
```
