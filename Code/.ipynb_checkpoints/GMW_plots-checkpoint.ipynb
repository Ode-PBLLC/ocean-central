```python
import pandas as pd
import matplotlib.pyplot as plt

country_dict = pd.read_csv("../Data/country_to_continent.csv") # download from https://github.com/lukes/ISO-3166-Countries-with-Regional-Codes/blob/master/all/all.csv
country_stats = pd.read_excel('../Data/gmw_v3_country_statistics_ha.xlsx') # download from https://zenodo.org/records/6894273

# Calculate the difference between the 1996 column and the rest of the columns
country_stats.iloc[:, 2:] = country_stats.iloc[:, 2:].sub(country_stats.iloc[:, 2], axis=0)

joined_df = country_stats.merge(country_dict[['region','alpha-3']],left_on="C_ID",right_on="alpha-3").groupby('region').sum().iloc[:, 2:]

joined_df = joined_df/1000

joined_df
```

```python
country_stats.query("Name == 'Global (ha)'").T.iloc[2:]
```

```python
import matplotlib.cm as cm

global_df = country_stats.query("Name == 'Global (ha)'").T.iloc[2:]/1000

# Plot the data with customized line colors
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)
global_df.plot(kind='line', ax=ax, colormap=cm.coolwarm_r, lw=2, legend=False)  # Use coolwarm colormap and thicker lines
plt.title('Total Change in Mangrove Area relative to 1996', fontname='Avenir')
plt.xlabel('Year', fontname='Avenir')
plt.ylabel('Total Change in Mangrove Area [1000 ha]', fontname='Avenir')
plt.xticks(rotation=45, fontname='Avenir')
plt.grid(True, linestyle='--', alpha=0.7)  # Add dashed grid lines with reduced transparency
plt.tight_layout()
plt.gca().set_facecolor('#f0f0f0')  # Set background color to light grey

plt.show()
```

```python
import matplotlib.cm as cm

# Define custom colormap
colors = cm.coolwarm_r((joined_df.iloc[:, -1] - joined_df.iloc[:, -1].min()) / (joined_df.iloc[:, -1].max() - joined_df.iloc[:, -1].min()))

# Plot the data with customized line colors
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)
joined_df.T.plot(kind='line', ax=ax, colormap=cm.coolwarm_r, lw=2, legend=False)  # Use coolwarm colormap and thicker lines
plt.title('Total Change in Mangrove Area by Region (1996-2020)', fontname='Avenir')
plt.xlabel('Year', fontname='Avenir')
plt.ylabel('Total Change in Mangrove Area [1000 ha]', fontname='Avenir')
plt.xticks(rotation=45, fontname='Avenir')
plt.grid(True, linestyle='--', alpha=0.7)  # Add dashed grid lines with reduced transparency
plt.tight_layout()
plt.gca().set_facecolor('#f0f0f0')  # Set background color to light grey

# Color lines based on 2020 values
for i, line in enumerate(ax.get_lines()):
    line.set_color(colors[i])

# Reorder legend based on 2020 values and flip the order
handles, labels = ax.get_legend_handles_labels()
sorted_labels = [label for _, label in sorted(zip(joined_df.iloc[:, -1], labels), reverse=True)]
sorted_handles = [handle for _, handle in sorted(zip(joined_df.iloc[:, -1], handles), reverse=True)]
plt.legend(sorted_handles, sorted_labels, prop={'family': 'Avenir'})

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

new_df = country_stats.merge(country_dict[['region','alpha-3']],left_on="C_ID",right_on="alpha-3").query("region == 'Africa'")

# Sort the DataFrame by the '2020' column
new_df_sorted = new_df.sort_values(by=2020)

# Calculate the maximum absolute value in the '2020' column
max_abs_value = 1000

# Define custom colormap with center at 0
colors = cm.coolwarm_r((new_df_sorted[2020] + max_abs_value) / (2 * max_abs_value))

# Plot the data
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)
new_df_sorted.plot(kind='barh', x='Name', y=2020, color=colors, ax=ax, legend=False)
plt.title('Total Change in Mangrove Area by Country (2020)', fontname='Avenir')
plt.xlabel('Total Change in Mangrove Area [ha]', fontname='Avenir')
plt.ylabel('',fontname='Avenir')
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add dashed grid lines with reduced transparency

# Add colorbar centered at 0
norm = plt.Normalize(-max_abs_value, max_abs_value)
sm = plt.cm.ScalarMappable(cmap=cm.coolwarm, norm=norm)
sm.set_array([])

plt.tight_layout()
plt.gca().set_facecolor('#f0f0f0')  # Set background color to light grey
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Assuming 'country_stats' and 'country_dict' are already defined and imported
country_stats = pd.read_excel('../Data/gmw_v3_country_statistics_ha.xlsx') # download from https://zenodo.org/records/6894273
new_df = country_stats.merge(country_dict[['region','alpha-3']],left_on="C_ID",right_on="alpha-3")

# Assuming 'max_abs_value' is already defined

# Sort the DataFrame by the '2020' column
new_df_sorted = new_df.sort_values(by=2020)

# Define minimum and maximum color values
min_color_value = 1e6
max_color_value = 3e6

# Normalize the data to the range of the colormap
norm = plt.Normalize(min_color_value, max_color_value)

# Define custom colormap with center at 0
colors = cm.RdYlGn_r(norm(new_df_sorted[2020]))

# Plot the data
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)
bars = new_df_sorted.tail(20).plot(kind='barh', x='Name', y=2020, color=colors, ax=ax, legend=False)
plt.title('Total Mangrove Area by Country (2020)', fontname='Avenir')
plt.xlabel('Total Mangrove Area [ha]', fontname='Avenir')
plt.ylabel('',fontname='Avenir')
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add dashed grid lines with reduced transparency

# Add colorbar
sm = plt.cm.ScalarMappable(cmap=cm.RdYlGn_r, norm=norm)
sm.set_array([])

# Color the y-axis labels red where region is 'Africa'
for label in bars.get_yticklabels():
    if new_df_sorted.loc[new_df_sorted['Name'] == label.get_text(), 'region'].iloc[0] == 'Africa':
        label.set_color('red')

plt.tight_layout()
plt.gca().set_facecolor('#f0f0f0')  # Set background color to light grey
plt.show()
```

## Coral

```python
import geopandas as gpd
from shapely.geometry import MultiPolygon

# Load EEZs and seagrass geometries
eez_file = "~/Downloads/Intersect_EEZ_IHO_v4_2020/Intersect_EEZ_IHO_v4_2020.shp"
coral_file = "~/Downloads/14_001_WCMC008_CoralReefs2021_v4_1/01_Data/WCMC008_CoralReef2021_Py_v4_1.shp"

eez = gpd.read_file(eez_file)
coral = gpd.read_file(coral_file)

# Check the CRS (Coordinate Reference System) and reproject if necessary
print("EEZ CRS:", eez.crs)
print("Seagrass CRS:", coral.crs)

# Reproject to a CRS that uses meters as the unit of measurement
# Let's use World Mercator projection (EPSG:3395) as an example
eez = eez.to_crs(epsg=3395)
coral = coral.to_crs(epsg=3395)

# Calculate intersection
intersection = gpd.overlay(eez, coral, how='intersection')

# Dissolve by country to get unique geometries per country
dissolved_intersection = intersection.dissolve(by='ISO_SOV1')

# Calculate area of dissolved geometries
dissolved_intersection['area_overlap'] = dissolved_intersection['geometry'].area

# Print the result
intersection_by_country = dissolved_intersection.reset_index()
intersection_by_country.sort_values('area_overlap')[['ISO_SOV1', 'area_overlap']]
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Assuming 'country_stats' and 'country_dict' are already defined and imported

# Assuming 'max_abs_value' is already defined

# Sort the DataFrame by the '2020' column
new_df_sorted = intersection_by_country.sort_values(by="area_overlap").merge(country_dict, left_on = "ISO_SOV1", right_on = "alpha-3")

new_df_sorted['area_overlap'] = new_df_sorted['area_overlap']/1e6

# Define minimum and maximum color values
min_color_value = 1e3
max_color_value = 1e5

# Normalize the data to the range of the colormap
norm = plt.Normalize(min_color_value, max_color_value)

# Define custom colormap with center at 0
colors = cm.RdYlGn_r(norm(new_df_sorted['area_overlap']))

# Plot the data
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)
bars = new_df_sorted.tail(20).plot(kind='barh', x='name', y='area_overlap', color=colors, ax=ax, legend=False)
plt.title('Total Coral Area by Country (2020)', fontname='Avenir')
plt.xlabel('Total Coral Area [$km^{2}$]', fontname='Avenir')
plt.ylabel('',fontname='Avenir')
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add dashed grid lines with reduced transparency

# Add colorbar
sm = plt.cm.ScalarMappable(cmap=cm.RdYlGn_r, norm=norm)
sm.set_array([])

# Color the y-axis labels red where region is 'Africa'
for label in bars.get_yticklabels():
    if new_df_sorted.loc[new_df_sorted['name'] == label.get_text(), 'region'].iloc[0] == 'Africa':
        label.set_color('red')

plt.tight_layout()
plt.gca().set_facecolor('#f0f0f0')  # Set background color to light grey
plt.show()
```
