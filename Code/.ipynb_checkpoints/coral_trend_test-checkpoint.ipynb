```python
import pandas as pd

df = pd.read_csv("/Users/pjs156/Downloads/AIMS_LTMP_manta-tow-by-reef/manta-tow-by-reef/manta-tow-by-reef.csv")
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out REEF_IDs without a value for REPORT_YEAR == 2010
reef_ids_2010 = df[df['REPORT_YEAR'] == 1995]['REEF_ID'].unique()
df_filtered = df[df['REEF_ID'].isin(reef_ids_2010)]

# Calculate the percent difference relative to the 2010 value for each REEF_ID
df_2010 = df_filtered[df_filtered['REPORT_YEAR'] == 1995].set_index('REEF_ID')['MEAN_LIVE_CORAL']
df_filtered['PERCENT_DIFF'] = df_filtered.apply(lambda row: ((row['MEAN_LIVE_CORAL'] - df_2010[row['REEF_ID']]) / df_2010[row['REEF_ID']]) * 100, axis=1)

# Calculate the mean percent difference for each REPORT_YEAR
mean_percent_diff = df_filtered.groupby('REPORT_YEAR')['PERCENT_DIFF'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=mean_percent_diff, x='REPORT_YEAR', y='PERCENT_DIFF', marker='o')

# Customizing the plot
plt.title('Mean Percent Difference in Live Coral Relative to 2010')
plt.xlabel('Report Year')
plt.ylabel('Mean Percent Difference')
plt.grid(True)

# Show plot
plt.show()
```

```python
combined_dataset.lon.min()
```

```python
import xarray as xr
import matplotlib.pyplot as plt
from PIL import Image
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np

# Extract the variable you want to plot
variable = combined_dataset['z_thresh']

# Set up figure and axis
fig, ax = plt.subplots(figsize=(8, 12), dpi=150, subplot_kw={'projection': ccrs.PlateCarree()})
plt.close()  # Close the figure so it doesn't display

# Initialize an empty list to store frames
frames = []

# Loop through each time step
for t in range(len(variable.time)):
    ax.clear()
    ax.set_aspect('auto')  # Set the aspect to auto to handle projection
    ax.add_feature(cfeature.LAND, edgecolor='white')
    ax.add_feature(cfeature.COASTLINE, edgecolor='white')
    ax.add_feature(cfeature.BORDERS, linestyle=':', edgecolor='white')
    
    # Plot a navy blue background
    ax.imshow(np.full((100, 100, 3), [0, 0, 128], dtype=np.uint8), 
              extent=[combined_dataset.lon.min().values, combined_dataset.lon.max().values, combined_dataset.lat.min().values, combined_dataset.lat.max().values], transform=ccrs.PlateCarree(), zorder=0)
    
    # Plot the variable data
    im = variable.isel(time=t).plot(ax=ax, cmap='cividis', vmin=0, vmax=30, add_colorbar=False, transform=ccrs.PlateCarree(), zorder=1)
    
    # Add title
    ax.set_title(f"Time step: {variable.time[t].values}", fontsize=12, pad=20)
    ax.set_xlabel("")
    ax.set_ylabel("")
    
    # Add colorbar
    if t == 0:
        cb = fig.colorbar(im, ax=ax, orientation='vertical', pad=0.05)
        cb.ax.tick_params(labelsize=10, color='white')  # Adjust colorbar tick font size and color
        cb.outline.set_edgecolor('white')  # Set the edge color of the colorbar to white
        plt.setp(plt.getp(cb.ax.axes, 'yticklabels'))  # Set the color of the colorbar labels to white

    # Create a PIL image from the plot
    fig.canvas.draw()
    img = Image.frombytes('RGB', fig.canvas.get_width_height(), fig.canvas.tostring_rgb())

    # Append the image to the frames list
    frames.append(img)

# Save the frames as an animated GIF using PIL
frames[0].save('animation.gif', save_all=True, append_images=frames[1:], optimize=False, duration=200, loop=0)
```
